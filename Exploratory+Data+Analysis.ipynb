```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6

data = pd.read_csv('C:/Users/JP/Documents/School/Practicum/Github/Practicum/data/dataset.csv')
```

```python
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'],format='%Y-%m-%d')
df = df.set_index('Date')
df.head()
```

```python
#Summary of data
df.describe()
```

```python
# Correlation of all features
df.corr()
```

```python
# Correlation of all features using Spearman
df.corr('spearman')
```

```python
# Correlation of all features using Kendall
df.corr('kendall')
```

```python
from bokeh.charts import Scatter, show
from bokeh.io import output_notebook

# Construct the scatter plot Tbonds and ETHOpen
p = Scatter(df, x='ETHOpen', y='TBondsOpenValue', title="ETH v Tbonds",
            xlabel="ETHOpen", ylabel="Tbonds")


output_notebook()

show(p)
```
